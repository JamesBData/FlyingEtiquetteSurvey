```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/Users/master/Downloads/flying-etiquette.csv')
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Age', data=data, palette='viridis')
plt.title('Percent Breakdown of Survey Participants by Age Range')
plt.xlabel('Age Range')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Is it rude to ask someone to switch seats with you in order to be closer to friends?', data=data, palette='pastel')
plt.title('Is it rude to ask someone to switch seats with you in order to be closer to friends?')
plt.xlabel('Feelings about Asking to Switch Seats')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Is itrude to ask someone to switch seats with you in order to be closer to family?', data=data, palette='pastel')
plt.title('Is itrude to ask someone to switch seats with you in order to be closer to family?')
plt.xlabel('Feelings about Asking to Switch Seats')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(x='Age', hue='Is itrude to ask someone to switch seats with you in order to be closer to family?', data=data, palette='Set2')
plt.title('Age Dimension Analysis for Asking to Switch Seats')
plt.xlabel('Age Range')
plt.ylabel('Count')
plt.legend(title='Asking to Switch Seats to be Closer to Family')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(x='Age', hue='Is it rude to ask someone to switch seats with you in order to be closer to friends?', data=data, palette='Set2')
plt.title('Age Dimension Analysis for Asking to Switch Seats')
plt.xlabel('Age Range')
plt.ylabel('Count')
plt.legend(title='Asking to Switch Seats to be Closer to Friends')
plt.show()
```
